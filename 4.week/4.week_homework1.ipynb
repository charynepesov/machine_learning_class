#### 1. Create a figure with 4 subplots.
- A. Histogram of Ages of Titanic
- B. Scatterplot of Age vs Fares
- C. Boxplot of Ages of Survivors vs Deceased
- D. Boxplot of Ages of Men vs Women

```python
import pandas as pd
import matplotlib.pyplot as plt
train = pd.read_csv('train.csv')
```

```python
mean = train.mean()
train.fillna(value=mean, inplace=True)
```

```python
fig = plt.figure(figsize = (40,40))
# fig.add_subplot(rows, cols, index)
a = fig.add_subplot(111)
a.hist(train['Age'])
a.set_title('Histogram of Ages of Titanic')

b = fig.add_subplot(121)
b.scatter(train['Age'], train['Fare'])
b.set_title('Scatterplot of Age vs Fares')

c = fig.add_subplot(211)
dec = train.Age[train.Survived==0]
sur = train.Age[train.Survived==1]
c.boxplot([dec, sur])
c.set_title('Boxplot of Ages of Survivors vs Deceased')

d = fig.add_subplot(221)
men = train.Age[train.Sex=='male']
women = train.Age[train.Sex=='female']
d.boxplot([men, women])
d.set_title('Boxplot of Ages of Men vs Women')
```

```python
men = train.Age[train.Sex=='male']
women = train.Age[train.Sex=='female']
_fig = plt.figure(figsize =(8, 5))
plt.boxplot([men, women])
plt.show()
```
